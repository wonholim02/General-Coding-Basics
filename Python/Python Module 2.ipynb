```python
import bs4
```

```python
req = requests.get('https://www.naver.com/')
naver_html = req.text
```

```python
soup = bs4.BeautifulSoup(naver_html,'html.parser')
```

```python
print(soup)
```

```python
soup.find('li',attrs={'class':'nav_item'})
```

```python
soup.find('li',attrs={'class':'nav_item'}).text
```

```python
soup.find_all('li',attrs={'class':'nav_item'})
```

```python
for i in soup.find_all('li',attrs={'class':'nav_item'}):
    print(i.text)
```

```python
req = requests.get('https://movie.naver.com/movie/sdb/rank/rmovie.nhn')
movie_html = req.text
soup = bs4.BeautifulSoup(movie_html,'html.parser')
temp = soup.find('div',attrs={'class':'tit3'}).text
print(temp.strip())
```

```python
count = 1
for i in soup.find_all('div',attrs = {'class':'tit3'}):
    print(count,"위 : ", i.text.strip())
    count += 1
```
